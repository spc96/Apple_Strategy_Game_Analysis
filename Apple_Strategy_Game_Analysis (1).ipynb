```python
import os

import numpy as np
from matplotlib import pyplot as plt
import matplotlib.dates as mdates

import string

import nltk

from operator import itemgetter

import pandas as pd

import datetime

import seaborn as sns
```

```python
data = pd.read_csv('appstore_games.csv', encoding = "ISO-8859-1")
```

```python
description = data["Description"]
keywords = []


for i in range(len(description)):
    s = description[i].lower().replace("\\n",' ').translate(str.maketrans('', '', string.punctuation)).split()
    keywords.append(s)

#print(keywords[0:5])
```

```python
flat_keywords = [item for sublist in keywords for item in sublist]

keywords_set = set(flat_keywords)

counts = dict()
for i in flat_keywords:
    counts[i] = counts.get(i, 0) + 1
```

```python
counts
```

```python
ordered_counts = sorted(counts, key=counts.get, reverse=True)
```

```python
top1000 = []
for i in range(1000):
    #print(ordered_counts[i],counts[ordered_counts[i]])
    top1000.append(ordered_counts[i])
```

```python
keywords_str = " ".join(keywords_set)

nouns = [word for (word, pos) in nltk.pos_tag(nltk.word_tokenize(keywords_str)) if pos[0] == 'N']
```

```python
nouns = set(nouns)
top1000 = set(top1000)
top_nouns = top1000.intersection(nouns)
```

```python
top_nouns

top_nouns_counts= []

for i in top_nouns:
    top_nouns_counts.append((i,counts[i]))
```

```python


top_nouns_counts = sorted(top_nouns_counts,key=itemgetter(1),reverse=True)

top_nouns_counts
```
